## Landmark: Great Falls Park

A National Park on the bank of the Potomac River. A popular outdoor area due to its complex of rapids and waterfalls, Great Falls is in reality situated between Virginia and Maryland. Great Falls has a long history - from Native American petroglyphs to man-made canals funded by President George Washington.
