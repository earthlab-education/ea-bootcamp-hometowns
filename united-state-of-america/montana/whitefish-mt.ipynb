## Glacier National Park

Some of the signs around Whitefish, MT claim that the town is the true gateway into Glacier National Park.  This park is absolutely stunning and a must visit on every bucket list.  The town of Whitefish sits at the southern edge of the park and allow direct access for visitors to drive the Going-to-the-Sun road which offers picturesque and sweeping views of the mountains and colorful valleys in the park.
